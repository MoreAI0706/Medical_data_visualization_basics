```python
import pandas as pd

# Sample medical data
data = {
'Name': ['John', 'Alice', 'Bob', 'Daisy', 'Tom'],
'Age': [29, 34, 40, 42, 31],
'BP': [120, 130, 140, 125, 125],
'Smoker': ['No', 'Yes', 'No', 'Yes', 'No']
 }

df = pd.DataFrame(data)
df
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.countplot(x='Smoker', data=df)
plt.title('🚬 Smoker Count')
plt.show()
```

```python
# Count values
smoker_counts = df['Smoker'].value_counts()

# Pie chart
plt.pie(smoker_counts, labels=smoker_counts.index, autopct='%1.1f%%', startangle=90)
plt.title('Smoker Distribution')
plt.axis('equal')  # Equal aspect ratio for a circle
plt.show()
```

```python
plt.hist(df['Age'], bins=5, color='skyblue', edgecolor='black')
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
sns.heatmap(df[['Age', 'BP']].corr(), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```

```python
sns.scatterplot(x='Age', y='BP', hue='Smoker', data=df)
plt.title('Age vs Blood Pressure')
plt.xlabel('Age')
plt.ylabel('Blood Pressure')
plt.show()
```
